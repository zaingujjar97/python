# **06 -Set and Its Methods(Mutable)**

1. Sets are unordered => Element's order doesn't matter
2. Sets are unindexed => Cannot access elements by index
3. There is no way to change items in sets.
4. Sets cannot contain duplicate values.
5. **Mutable** Aap set ke andar nayi cheezein add, remove, ya update kar sakte ho.

```python
s = {1, 2, 3, 4}
 
s = set()     # ‚úîÔ∏è Ye empty set banata hai      <-- Note: set() is used to create an empty set
print(s)      # Output: set()
print(type(s))  # Output: <class 'set'>

s.add(5)
print(s)                  # {1, 2, 3, 4, 5}
s.remove(2)
print(s)                  # {1, 3, 4, 5}
s.discard(10)             # No error even if 10 not present
print(s.pop())            # Random element remove karke return karega
print(s.union({5,6,7}))   # {1, 3, 4, 5, 6, 7}
print(s.intersection({3,4,10}))  # {3, 4}
print(s.difference({3}))          # Set ke wo elements jo 3 ke baghair hain
print({1,3}.issubset(s))          # True ya False
print(s.issuperset({1}))          # True ya False
s.clear()
print(s)                         # set()
```

# **08 ‚Äì Functions**

**Text Explanation:**

Function ek reusable code block hai jo specific task perform karta hai. Is se code modular aur organized rehta hai. Functions code ko modular aur reusable banate hain.
Yahan aap dekhenge ke function ke andar parameters, return statements, aur docstrings kaise use hote hain.

**Explanation:**

def factorial(n): se function define hota hai.
Docstring function ke purpose ko bayan karta hai.
For loop se factorial calculate hota hai aur result return hota hai.

**Example :**

```python
def factorial(n):
    """
    Is function ka maqsad n ka factorial calculate karna hai.
    Agar n 0 ho to factorial 1 return karta hai.
    """
    if n == 0:
        return 1
    else:
        result = 1
        # 1 se n tak multiplication
        for i in range(1, n + 1):
            result *= i
        return result

# Function ko call kar ke output dekhna
number = 5
print(f"{number} ka factorial hai:", factorial(number))
```

# **11 ‚Äì Object-Oriented Programming (OOP)**
**Text Explanation:**

OOP mein aap classes aur objects ka istemal karte hain. Is se aap complex programs ko chhote, manageable parts mein divide kar sakte hain. Concepts mein inheritance aur method overriding shamil hain.


**‚≠êSimple Class and Object Example**

```python
class person:
  name = "zain gujjar"
  professoinal = " agentic ai developer"
  language = "english"
  salary = 230000             # --->  this is class attribute
  age = "21"
  professoinal = "developer"
  def get_info(self):
    print(f"{self.name} is using {self.language} and it salary is {self.salary} and it age is  {self.age}")
  def decorator(self):
    print(f"GOOD MORINING {self.name} ")

# OBJECTS 1 OF THE CLASS
zain = person()
zain.decorator()
zain.name = "zain gujjar"     # ---> instance attribute
zain.professoinal = " agentic ai developer"
zain.language = "python"
zain.get_info()
print(zain.name,zain.language,zain.salary,zain.age, zain.professoinal) 

# OBJECTS 2 OF THE CLASS
almas = person()
almas.decorator()
almas.name = "illustrator"    # ---> instance attribute/object attribute
almas.professoinal = " graphic desinger"
almas.salary = 100000
almas.get_info()
print(almas.name,almas.language,almas.salary,almas.age, almas.professoinal)
```

**‚≠êOOPS Inheritance Example**

```python
# üêæ Base class
class Animal:
    def __init__(self, name):  # Constructor
        self.name = name        # Instance Attribute

    def speak(self):            # Method
        return "Koi khas sound nahi."  # Default sound


# üê∂ Derived class: Dog (inherits from Animal)
class Dog(Animal):  # Inheritance
    def speak(self):  # Method override (polymorphism)
        return f"{self.name} bolta hai: Bark!"


# üê± Derived class: Cat (inherits from Animal)
class Cat(Animal):  # Inheritance
    def speak(self):  # Method override (polymorphism)
        return f"{self.name} bolta hai: Meow!"


# ‚úÖ Creating objects of child classes
dog1 = Dog("Max")
cat1 = Cat("Mia")

# ‚úÖ Calling speak method (method overriding in action)
print(dog1.speak())  # Output: Max bolta hai: Bark!
print(cat1.speak())  # Output: Mia bolta hai: Meow!
```

**‚≠êClass Method**

```python
class Test:
    a = 1  # class attribute

# ‚≠ê jab @classmethod use hota ha to class attribute chalta ha or instance attribute ni chalta                  <-- very important point
    @classmethod 
    def show(cls):                           # cls refers to the class, not instance
        print(f"The class attribute is {cls.a}")

p = Test()
p.a = 35  # üëà This is an instance attribute (will NOT affect class method)

p.show()
```

**‚≠êStatic Method**

```python
# self and cls are used in instance and class methods respectively.
class Math:
    @staticmethod
    def add(a, b):
        return a + b

print(Math.add(3, 4))  # ‚úÖ No object or self/cls needed
```

**‚≠êSuper Method**

```python
# Used in inheritance to call parent class‚Äôs method and run __init_ method of the parent class
class employ():
  def __init__(self) :
     print("the constructor of the employ")
  a = 1

class programmer(employ):
    def __init__(self) :
      print("the constructor of the programmer")
    b = 2
class manager(programmer,employ):
    def __init__(self) :
      super().__init__()     # --> used super and call the all __init_ method of the parent class
      print("the constructor of the manager")
    c = 3

p = employ()
print(p.a)

p = programmer()
print(p.a,p.b)

p = manager()
print(p.a,p.b,p.c)
```

# **14 ‚Äì Final Projects**

Text Explanation:

Is final session mein aap apni seekhi hui tamam skills ko aik chhote project mein integrate karte hain. Ye project real-world application jaisa hota hai jisme user input, data processing, error handling aur modular design shamil hoti hai.


 **Explanation:**


*   Menu Driven Program: User ko menu options diye gaye hain.
*   Functions: Har functionality (task add, view, remove) ko alag function mein define kiya gaya hai.

*   Looping & Control Flow: While loop continuously menu display karta hai jab tak user exit nahi karta.
*   Error Handling: User input ke errors ko handle karne ke liye try-except use hua hai.



```python
# Ek simple To-Do List application
def display_menu():
    print("\n==== To-Do List Menu ====")
    print("1. Task add karein")
    print("2. Tasks dekhain")
    print("3. Task remove karein")
    print("4. Exit")

def add_task(tasks):
    task = input("Naya task likhain: ")
    tasks.append(task)
    print(f"Task '{task}' add ho gaya hai.")

def view_tasks(tasks):
    if not tasks:
        print("Abhi tak koi task nahi.")
    else:
        print("\nAapke tasks:")
        for index, task in enumerate(tasks, start=1):
            print(f"{index}. {task}")

def remove_task(tasks):
    view_tasks(tasks)
    try:
        task_num = int(input("Remove karne ke liye task number daalein: "))
        if 1 <= task_num <= len(tasks):
            removed = tasks.pop(task_num - 1)
            print(f"Task '{removed}' remove ho gaya hai.")
        else:
            print("Invalid task number!")
    except ValueError:
        print("Please ek valid number daalein.")

def main():
    tasks = []
    while True:
        display_menu()
        choice = input("Apni choice (1-4) daalein: ")
        if choice == '1':
            add_task(tasks)
        elif choice == '2':
            view_tasks(tasks)
        elif choice == '3':
            remove_task(tasks)
        elif choice == '4':
            print("Program exit ho raha hai. Shukriya!")
            break
        else:
            print("Invalid choice, dobara try karein.")

if __name__ == "__main__":
    main()
```
