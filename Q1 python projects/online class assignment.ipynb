```python
import random

print("Welcome to the High-Low Game!")
print("-" * 100)
score = 0
num_round = 5
for round in range(1,num_round + 1):


    comp_num = random.randint(1,100)
    user_num = random.randint(1,100)

    print(f"ROUND {round}","-" * 25)

    print(f"Your number is {user_num}")
    user_inp =str( input("Do you think your number is higher or lower than the computer's? "))


    if (
        (user_inp == "higher" and user_num < comp_num)
        or
        (user_inp == "lower" and user_num > comp_num)
    ):

     print(f"You were right! The computer's number was {comp_num}, you got point +1.....")
     score += 1

    else:
     print(f"wrong, that's incorrect. The computer's number was {comp_num}")


print(f"your score is {score}")


if score == 5:
  print("Wow! You played perfectly!")
elif score > 3:
  print("Good job, you played really well!")
else:
  print("Better luck next time!")
```

```python
all_planets = {

    "Earth": 100,
    
    "Mercury": 37.6,

    "Venus": 88.9,

     "Mars": 37.8,

     "Jupiter": 236.0,

     "Saturn": 108.1,

     "Uranus": 81.5,

     "Neptune": 114.0,
}

def weight_on_planets():
    earth_weight = int(input("Enter your earth weight : "))
    fvrt_planet = input("Enter your favorite planet : ")
    if fvrt_planet in all_planets:
        weight = earth_weight * all_planets[fvrt_planet] / 100
        bounded = round(weight,2)
        print(f"your weight on {fvrt_planet} is {bounded}")
    else:
        print("please check the spelling of the planet")


weight_on_planets()
```

```python
class account:
    def __init__(self, blnc , acc ):
        self.balance = blnc
        self.account = acc

    def debit(self, amount):
        self.balance += amount
        print("YOU add", amount, "in your account")
        print("total balance = ", self.get_balance())

    def credit(self, amount):
        self.balance -= amount
        print("You Add" ,amount, "in your account")
        print("total balance = ", self.get_balance())

    def get_balance(self):
        return self.balance

acc1 = account(5 ,"pk4532")

acc1.debit(5)
acc1.credit(3)
acc1.credit(7)
acc1.credit(1)
```
