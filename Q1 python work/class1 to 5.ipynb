**List index**

```python
#shopping list
shoping_list : list[str] = [ "juice","milk","fruit","water","bread" ]

print(shoping_list [4])
print(shoping_list[2])

# Find the Lenght of the list
print(len(shoping_list))

# List slicing
print(shoping_list[2:5])
print(shoping_list)

#remove in list
shoping_list.remove("milk")
print(shoping_list)

# insert in list
shoping_list.insert(2, "sugar")
print(shoping_list)

#pop in list
shoping_list.pop(4)
print(shoping_list)
```
