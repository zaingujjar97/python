**Class**

```python
class bmw:
  def __init__(car,name,model,color,engine):
    car.name = name
    car.model = model
    car.color = color
    car.engine = engine

  def __str__(car):
    return f"Name: {car.name}\nModel: {car.model}\nColor: {car.color}\nEngine: {car.engine}"
  
 
car1 = bmw ("odi", 2004, "black", "v8")
print(car1)
```

```python
class Toyota:
    def __init__ ( self, markx, toyota, black, v6, model ):
        self.name = markx
        self.company = toyota
        self.colour = black
        self.engine = v6
        self.model = model

    def __str__(self):
        return f"Name: {self.name}\nCompany: {self.company}\nColour: {self.colour}\nEngine: {self.engine}\nModel: {self.model}"

mycar = Toyota("landcrusier","toyota","black","v6",2003)
print(mycar)
```

**Error Handling in Python**

```python
try:
    num = int(input("Enter the number = "))
    result = num / 10
    print(f"YOUR RESULT: {result}")

except ZeroDivisionError:
    print("Error: You cannot divide by zero!")

except ValueError:
    print("Invalid input! Please enter a number.")

except Exception as e:
    print(f"Some unexpected error occurred:")
```

**Condition VS Range** ✅

```python
for number in range(5):  # This generates numbers from 0 to 4 (not including 5)
    print(number)



age = 18
if age >= 18:  # This condition checks if age is 18 or greater
    print("You are an adult.")
```

**While Loop** ✅

```python
time = 10

while time >= 1 :
  print(f"start {time}...")
  time -= 1

print("GOOD NIGHT")
```
